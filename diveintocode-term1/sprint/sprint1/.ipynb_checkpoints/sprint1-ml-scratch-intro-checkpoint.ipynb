## 【問題1】クロスバリデーション
事前学習期間は検証用データを分割しておき、それに対して指標値を計算することで検証を行っていました。しかし、分割の仕方により精度は変化します。実践的には クロスバリデーション を行います。

具体的には分割を複数回行い、それぞれに対して学習と検証を行う方法です。複数回の分割を行う関数はscikit-learnにKFoldとして用意されています。

[sklearn.model_selection.KFold — scikit-learn 0.20.2 documentation](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html#sklearn.model_selection.KFold)

### データ読み込み

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 200)

# read train and test data
df = pd.read_csv('application_train.csv')
```

```python
df = df.iloc[:30000]
#Confirm data
print(df.shape)
df.head()
```

### One-hot Encoding

```python
#カテゴリカル変数をOneHotEncodingを行う
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

```python
#trainの欠損値を確認
df_dummies.isnull().sum()
```

```python
# 欠損値を0で補完
df_dummies = df_dummies.fillna(0)
```

```python
#欠損がなくなっていることを確認
df_dummies.isnull().sum()
```

### 特徴量の抽出

```python
# trainデータのTARGETとfeatures間の相関係数を算出
feature_corr = df_dummies.corr(method='pearson')
feature_corr
```

```python
# 相関係数行列から目的変数の列を抽出
feature_target_corr = feature_corr.loc[:,'TARGET']
feature_target_corr
```

```python
feature_target_corr = abs(feature_target_corr).sort_values(ascending=False)
feature_target_corr
```

```python
# df_dummiesからTARGETと相関係数の高い特徴量のみを抽出
#一番相関大きいのはTARGET vs TARGETの1.0で抽出する意味がないので、0列目は除外
df_selected = df_dummies.loc[:, feature_target_corr[1:4].index]
#確認
df_selected.shape
```

```python
df_selected
```

### 分割

```python
from sklearn.model_selection import KFold

#generate instance
kf = KFold(n_splits=4, random_state=None, shuffle=False)

for train_index, test_index in kf.split(df_selected):
    print("train_index", train_index)
    print("test_index: ", test_index)
```

```python
# Pipeline クラスをインポート
from sklearn.pipeline import Pipeline
# StanardScalerクラスをインポート
from sklearn.preprocessing import StandardScaler

# 機械学習モデルをインポート
from sklearn.svm import SVC
#評価のライブラリ
from sklearn.metrics import classification_report
#ROC曲線描画用にインポート
from sklearn.metrics import  roc_curve
import  matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score

def SVC_classify(X_train, X_test, y_train, y_test):

    #パイプライン作成
    pl = Pipeline([
       ('scl', StandardScaler()),
        ('svc', SVC(probability=True))
    ])

    #train with train data
    pl.fit(X_train, y_train)
    
    #Estimate test value
    pred = pl.predict(X_test)
    prob = pl.predict_proba(X_test)
    
    #Evaluation
    print(classification_report(y_test, pred))
    
    #ROC
    fpr, tpr, thresholds = roc_curve(y_test, prob[:, 1])
    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr)
    plt.title("ROC curve")
    plt.xlabel("False Positive Ratio")
    plt.ylabel("True Positive Ratio")
    plt.show()
    
    #AUC
    auc = roc_auc_score(y_test, pred)
    print("AUC: ", auc)
    return pred, prob
```

```python
for train_index, test_index in kf.split(df_selected, df.loc[:, 'TARGET']):
    SVC_classify(df_selected.loc[train_index], df_selected.loc[test_index], df.loc[train_index, 'TARGET'], df.loc[test_index, 'TARGET'])
```

```python
from sklearn.model_selection import KFold
k_fold = KFold(n_splits=4, random_state=None, shuffle=False)
```

```python
# Pipeline クラスをインポート
from sklearn.pipeline import Pipeline
# StanardScalerクラスをインポート
from sklearn.preprocessing import StandardScaler

# 機械学習モデルをインポート
from sklearn.svm import SVC
#評価のライブラリ
from sklearn.metrics import classification_report
#ROC曲線描画用にインポート
from sklearn.metrics import  roc_curve
import  matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score

C_list = np.logspace(-9, 0, 10)
score = np.zeros((len(C_list),3))
tmp_train, tmp_test = list(), list()
i = 0
for C in C_list:
    print(i)
    svc = SVC(C=C, kernel='rbf', gamma=0.001)
    for train, test in k_fold.split(df_selected):
        print(train, test)
        svc.fit(df_selected.iloc[train, :], df.loc[train, 'TARGET'])
        tmp_train.append(svc.score(df_selected.iloc[train], df.loc[train, 'TARGET']))
        tmp_test.append(svc.score(df_selected.iloc[test], df.loc[test, 'TARGET']))
        score[i,0] = C
        score[i,1] = sum(tmp_train) / len(tmp_train)
        score[i,2] = sum(tmp_test) / len(tmp_test)
        del tmp_train[:]
        del tmp_test[:]
    i = i + 1
```

```python
xmin, xmax = score[:,0].min(), score[:,0].max()
ymin, ymax = score[:,1:2].min()-0.1, score[:,1:2].max()+0.1
plt.semilogx(score[:,0], score[:,1], c = "r", label = "train")
plt.semilogx(score[:,0], score[:,2], c = "b", label = "test")
plt.axis([xmin,xmax,ymin,ymax])
plt.legend(loc='upper left')
plt.xlabel('C')
plt.ylabel('score')
plt.show
```

Cが増えるにつれて処理時間が指数関数的に長くなる?

```python
g_list = np.logspace(-2, 4, 7) 
score = np.zeros((len(g_list),3))
tmp_train, tmp_test = list(), list()
i = 0
for gamma in g_list:
    svc = SVC(C=1, gamma=gamma, kernel='rbf')
    print(i)
    for train, test in k_fold.split(df_selected):
        print(train, test)
        svc.fit(df_selected.iloc[train, :], df.loc[train, 'TARGET'])
        tmp_train.append(svc.score(df_selected.iloc[train], df.loc[train, 'TARGET']))
        tmp_test.append(svc.score(df_selected.iloc[test], df.loc[test, 'TARGET']))
        score[i,0] = gamma
        score[i,1] = sum(tmp_train) / len(tmp_train)
        score[i,2] = sum(tmp_test) / len(tmp_test)
        del tmp_train[:]
        del tmp_test[:]
    i = i + 1
```

```python
xmin, xmax = score[:,0].min(), score[:,0].max()
ymin, ymax = score[:,1:2].min()-0.1, score[:,1:2].max()+0.1
plt.semilogx(score[:,0], score[:,1], c = "r", label = "train")
plt.semilogx(score[:,0], score[:,2], c = "b", label = "test")
plt.axis([xmin,xmax,ymin,ymax])
plt.legend(loc='upper left')
plt.xlabel('gamma')
plt.ylabel('score')
plt.show
```

gammaに比例して処理時間が長くなる
