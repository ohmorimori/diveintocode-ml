## 【問題1】クロスバリデーション
事前学習期間は検証用データを分割しておき、それに対して指標値を計算することで検証を行っていました。しかし、分割の仕方により精度は変化します。実践的には クロスバリデーション を行います。

```python
# Pipeline クラスをインポート
from sklearn.pipeline import Pipeline
# StanardScalerクラスをインポート
from sklearn.preprocessing import StandardScaler

# 機械学習モデルをインポート
from sklearn.svm import SVC
#評価のライブラリ
from sklearn.metrics import classification_report
#ROC曲線描画用にインポート
from sklearn.metrics import  roc_curve
import  matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score

C_list = np.logspace(-9, 0, 10)
score = np.zeros((len(C_list),3))
tmp_train, tmp_test = list(), list()
i = 0
start = time.time()#開始時刻
for C in C_list:
    print(i)
    lap = time.time()#lap time
    print("lap: ", lap-start, 'sec')
    svc = SVC(C=C, kernel='rbf', gamma=0.001)
    for train, test in k_fold.split(df_selected):
        print(train, test)
        svc.fit(df_selected.iloc[train, :], df.loc[train, 'TARGET'])
        tmp_train.append(svc.score(df_selected.iloc[train], df.loc[train, 'TARGET']))
        tmp_test.append(svc.score(df_selected.iloc[test], df.loc[test, 'TARGET']))
        score[i,0] = C
        score[i,1] = sum(tmp_train) / len(tmp_train)
        score[i,2] = sum(tmp_test) / len(tmp_test)
        del tmp_train[:]
        del tmp_test[:]
    i = i + 1
```

```python
xmin, xmax = score[:,0].min(), score[:,0].max()
ymin, ymax = score[:,1:2].min()-0.1, score[:,1:2].max()+0.1
plt.semilogx(score[:,0], score[:,1], c = "r", label = "train")
plt.semilogx(score[:,0], score[:,2], c = "b", label = "test")
plt.axis([xmin,xmax,ymin,ymax])
plt.legend(loc='upper left')
plt.xlabel('C')
plt.ylabel('score')
plt.show
```

## 【問題2】グリッドサーチ
これまで分類器のパラメータは基本的にデフォルトの設定を使用していました。パラメータの詳細は今後のSprintで学んでいくことになりますが、パラメータは状況に応じて最適なものを選ぶ必要があります。パラメータを探索するために グリッドサーチ と呼ばれる総当たり的手法が一般的に利用されます。

グリッドサーチをパイプラインの中に組み込みましょう。

```python
#Cross validationのためにStratifiedKFold
from sklearn.model_selection import StratifiedKFold

#標準化のために StanardScaler
from sklearn.preprocessing import StandardScaler

# 機械学習モデルをインポート
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

#評価のライブラリ
from sklearn.metrics import classification_report
#ROC曲線描画用にインポート
from sklearn.metrics import  roc_curve

#時間を計測するためにtime関数を取得
import time
```

```python
def grid_search(model, X_train, X_test, y_train, y_test):
    
    #train with train data
    model.fit(X_train, y_train)
    
    #Estimate test value
    best_score = model.best_score_
    best_params = model.best_params_
    
    #Estimate test value
    pred = model.predict(X_test)
    prob = model.predict_proba(X_test)
    #AUC
    auc = roc_auc_score(y_test, prob[:, 1])
    
    #print("Classification report for classifier %s:\n%s\n" % (model, metrics.classification_report(y_test, pred)))
    print('Best Score: ', best_score)
    print('Best Params: ', best_params)
    print('AUC: ', auc)
```

```python
def grid_search_2(X, y):
    """
    Parameters
    -----
    X : DataFrame
    y : DataFrame
    
    Notes
    -----
    KFold（K分割）を行い、ロジスティック回帰を行う。
    また、グリッドサーチを用いることで、正則化項の適切なパラメータを探し出す。
    
    """
    k_fold = StratifiedKFold(n_splits=5, random_state=None, shuffle=False)
    

    start= time.time()#start time
    
    for train_index, test_index in k_fold.split(df_selected.loc[:, df_selected.columns != 'TARGET'], df_selected.loc[:, 'TARGET']):
        X_train = df_selected.loc[train_index, df_selected.columns != 'TARGET']
        X_test = df_selected.loc[test_index, df_selected.columns != 'TARGET']
        y_train = df_selected.loc[train_index, 'TARGET']
        y_test = df_selected.loc[test_index, 'TARGET']
        
        #standarize data
        scaler = StandardScaler()
        scaler.fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)
        
        #modelのインスタンス生成
        lgr = LogisticRegression()
        
        #GridSearchのパラメータ
        param_grid = {
            'C' : [0.001, 0.01, 0.1, 1, 10, 100],
            
        }
        #GridSearch
        grid_search = GridSearchCV(
            lgr,
            param_grid
        )
        
        #学習
        grid_search.fit(X_train, y_train)
        
        #推定
        pred = grid_search.predict(X_test)
        prob = grid_search.predict_proba(X_test)
        #評価
        #AUC
        auc_score = roc_auc_score(y_test, prob[:, 1])
        best_score = grid_search.best_score_
        best_params = grid_search.best_params_
        
        lap= time.time()#lap time
        
        print('-------------------------')
        print('AUC: ', auc_score)
        print('Best Score: ', best_score)
        print('Best Params: ', best_params)
        print('Processed Time: ', lap-start, 'sec')
        print('-------------------------\n')
```

```python
def grid_search_3(model, X, y, n_folds):
    """
    Parameters
    -----
    model : classification model
    X : DataFrame
    y : DataFrame
    n_folds : int
        number of folds
    
    Notes
    -----
    Stratified KFoldを行い、ロジスティック回帰を行う。
    また、グリッドサーチを用いることで、正則化項の適切なパラメータを探し出す。
    
    """
    k_fold = StratifiedKFold(n_splits=n_folds, random_state=None, shuffle=False)
    

    start= time.time()#start time
    
    for train_index, test_index in k_fold.split(df_selected.loc[:, df_selected.columns != 'TARGET'], df_selected.loc[:, 'TARGET']):
        X_train = df_selected.loc[train_index, df_selected.columns != 'TARGET']
        X_test = df_selected.loc[test_index, df_selected.columns != 'TARGET']
        y_train = df_selected.loc[train_index, 'TARGET']
        y_test = df_selected.loc[test_index, 'TARGET']
                
        #学習
        model.fit(X_train, y_train)
        
        #推定
        pred = model.predict(X_test)
        prob = model.predict_proba(X_test)
        
        #評価
        #AUC
        auc_score = roc_auc_score(y_test, prob[:, 1])        
        best_score = model.best_score_
        best_params = model.best_params_
        
        lap= time.time()#lap time
        
        print('-------------------------')
        print('AUC: ', auc_score)
        print('Best Score: ', best_score)
        print('Best Params: ', best_params)
        print('Processed Time: ', lap-start, 'sec')
        print('-------------------------\n')
```
